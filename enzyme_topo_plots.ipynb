# Plotting up old JGI data
Biochemical parameters vs. network measurements.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
bacteria = pd.read_csv('bacteria_topo_ecs.csv')
archaea = pd.read_csv('archaea_topo_ecs.csv')
eukarya = pd.read_csv('eukarya_topo_ecs.csv')
JGI = pd.read_csv('JGI_topo_ecs.csv')
YNP = pd.read_csv('YNP_topo_ecs.csv')
```

```python
df = pd.concat([bacteria, archaea, eukarya, JGI, YNP], sort=False)
```

```python
sns.scatterplot(x=bacteria['total enzymes'], y=bacteria['nbr_nodes'])
plt.show()

sns.scatterplot(x=bacteria['total enzymes'], y=bacteria['nbr_rxn'])
```
