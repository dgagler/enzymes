**Ascomycota** - sac fungi

**Basidiomycota** - filamentous fungi 

**Streptophyta** - land plants and green algae

**Apicomplexa** - obligate endoparasites of animals

**Chlorophyta** - green algae/seaweed

**Mucoromycota** - uncertain fungi

**Chytridiomycota** - zoosporic fungi

**Chordata** - vertebrates

**Microsporidia** - spore-forming fungi; unicellular parasites

**Zoopagomycota** - weird fungi?

**Bacillariophyta** - diatoms

**Porifera** - sponges

**Nematoda** - roundworms...nematodes

**Heterolobosea** - amoeba heterotroph things

**Perkinsozoa** - proposed classification, intracellular parasites


```python
### EUKARYA TAXA DISTRIBUTIONS - gross...but it works

### putting spaces in front of each item cuz that's how the data currently exists...need to fix
fungi = [' Ascomycota', ' Basidiomycota',' Mucoromycota', ' Chytridiomycota',
         ' Microsporidia', ' Zoopagomycota', ' Blastocladiomycota']
plants_algae = [' Streptophyta', ' Chlorophyta', ' Bacillariophyta', ' Cryptophyta']
parasites = [' Apicomplexa', ' Perkinsozoa']
animals = [' Chordata', ' Cnidaria', ' Annelida', ' Porifera', ' Placozoa', 
           ' Mollusca', ' Nematoda', ' Arthropoda']
protists = [' Ciliophora', ' Cercozoa', ' Perkinsozoa', ' Heterolobosea']

eukarya_phylum_dict = dict(eukarya_phylum_counts)

fungi_dict = {}

for k,v in eukarya_phylum_counts.items():
    if k in fungi:
        fungi_dict[k] = v
        
fungi_dict['total'] = sum(fungi_dict.values())

###############################################

green_dict = {}

for k,v in eukarya_phylum_counts.items():
    if k in plants_algae:
        green_dict[k] = v

green_dict['total'] = sum(green_dict.values())

###############################################

parasite_dict = {}

for k,v in eukarya_phylum_counts.items():
    if k in parasites:
        parasite_dict[k] = v
parasite_dict['total'] = sum(parasite_dict.values())

###############################################

animal_dict = {}

for k,v in eukarya_phylum_counts.items():
    if k in animals:
        animal_dict[k] = v

animal_dict['total'] = sum(animal_dict.values())

###############################################

protist_dict = {}

for k,v in eukarya_phylum_counts.items():
    if k in protists:
        protist_dict[k] = v
        
protist_dict['total'] = sum(protist_dict.values())

fungi_percent = round(fungi_dict['total'] / sum(eukarya_phylum_counts.values()) * 100, 2)
green_percent = round(green_dict['total'] / sum(eukarya_phylum_counts.values()) * 100, 2)
parasite_percent = round(parasite_dict['total'] / sum(eukarya_phylum_counts.values()) * 100, 2)
animal_percent = round(animal_dict['total'] / sum(eukarya_phylum_counts.values()) * 100, 2)
protist_percent = round(protist_dict['total'] / sum(eukarya_phylum_counts.values()) * 100, 2)
unclassified_percent = round(eukarya_phylum_dict[' unclassified'] / sum(eukarya_phylum_counts.values()) * 100, 2)

eukarya_sizes=[fungi_dict['total'], green_dict['total'], parasite_dict['total'],
              animal_dict['total'], protist_dict['total'], eukarya_phylum_dict[' unclassified']]
colors = ['black', 'darkmagenta', 'mediumvioletred', 'crimson', 'darkorange', 'gold']
eukarya_percents = ['{}%'.format(fungi_percent), '{}%'.format(green_percent), '{}%'.format(parasite_percent),
           '{}%'.format(animal_percent), '{}%'.format(protist_percent), '{}%'.format(unclassified_percent)]
eukarya_labels = ['Fungi', 'Plants/Algae', 'Parasites', 'Animals', 'Protists', 'Unclassified']


plt.pie(eukarya_sizes, colors=colors, labels=eukarya_percents)
plt.title('JGI Eukarya Genomes', fontsize=14)
plt.legend(eukarya_labels, loc='center left', bbox_to_anchor=(0.95, 0.75), fontsize=12)
plt.tight_layout()
plt.savefig('/Users/dgagler/School/SecondaryProject/RedoxEnzymes/JGI_2019/plots/eukarya_phylum_pie')
```

```python
archaea_phyla_dict = {}

count = 0
for k,v in archaea_phylum_counts.items():
    if v >= 40:
        archaea_phyla_dict[k] = v
    if v < 40:
        count += v
archaea_phyla_dict['other'] = count

archaea_phyla_dict
```

```python
archaea_sizes = list(archaea_phyla_dict.values())
archaea_labels = list(archaea_phyla_dict.keys())
archaea_percents = []

archaea_total = sum(archaea_sizes)
for i in archaea_sizes:
    percent = round((i/archaea_total) * 100, 2)
    archaea_percents.append('{}%'.format(percent))
```

```python
plt.pie(archaea_sizes, colors=colorss, labels=archaea_percents)
plt.title('JGI Archaea Genomes', fontsize=14)
plt.legend(archaea_labels, loc='center left', bbox_to_anchor=(0.95, 0.75), fontsize=12)
plt.tight_layout()
plt.savefig('/Users/dgagler/School/SecondaryProject/RedoxEnzymes/JGI_2019/plots/archaea_phylum_pie')
```

```python
bacteria_phyla_dict = {}

count = 0
for k,v in bacteria_phylum_counts.items():
    if v >= 200:
        bacteria_phyla_dict[k] = v
    if v < 200:
        count += v
bacteria_phyla_dict['other'] = count

bacteria_phyla_dict
```

```python
bacteria_sizes = list(bacteria_phyla_dict.values())
bacteria_labels = list(bacteria_phyla_dict.keys())
bacteria_percents = []

bacteria_total = sum(bacteria_sizes)
for i in bacteria_sizes:
    percent = round((i/bacteria_total) * 100, 2)
    bacteria_percents.append('{}%'.format(percent))
```
